# Repetition code data and analysis

```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
cmap = plt.colormaps['gist_rainbow']

colors = cmap(np.linspace(0,1,20))
```

## Open data

```python
with open("data/bit_flip_rep_code.p", "rb") as input_file:
    bit = pickle.load(input_file)
with open("data/phase_flip_rep_code.p", "rb") as input_file:
    phase = pickle.load(input_file)
```

## Plot data

```python
color_list = [plt.get_cmap("tab10").colors[i] for i in range(10)] 

x = np.array([1,2,3])
y= np.array([2,2,2])
plt.errorbar(x,y, yerr=None, marker='s', color='k', linestyle='dotted', label='phase flip code')
plt.errorbar(x,y, yerr=None, marker='o', color='k', linestyle='dashed', label='bit flip code')
plt.errorbar(x,y, yerr=None,  color=color_list[0], label='SE = 1')
plt.errorbar(x,y, yerr=None,  color=color_list[2], label='SE = 3')
plt.errorbar(x,y, yerr=None,  color=color_list[4], label='SE = 5')
plt.errorbar(x,y, yerr=None,  color=color_list[6], label='SE = 7')
plt.errorbar(x,y, yerr=None,  color=color_list[8], label='SE = 9')


plt.errorbar(list(bit[0][1].keys()),list(bit[0][1].values()),yerr=[list(bit[1][1].values()),list(bit[2][1].values())], label='_nolegend_', marker='o', alpha=1.0, color=color_list[0], linestyle='dashed',markersize=5, capsize=3, ecolor=color_list[0], markerfacecolor=color_list[0], markeredgecolor=color_list[0])
plt.errorbar(list(bit[0][3].keys()),list(bit[0][3].values()),yerr=[list(bit[1][3].values()),list(bit[2][3].values())], label='_nolegend_', marker='o', alpha=1.0, color=color_list[2], linestyle='dashed',markersize=5, capsize=3, ecolor=color_list[2], markerfacecolor=color_list[2], markeredgecolor=color_list[2])
plt.errorbar(list(bit[0][5].keys()),list(bit[0][5].values()),yerr=[list(bit[1][5].values()),list(bit[2][5].values())], label='_nolegend_', marker='o', alpha=1.0, color=color_list[4], linestyle='dashed',markersize=5, capsize=3, ecolor=color_list[4], markerfacecolor=color_list[4], markeredgecolor=color_list[4])
plt.errorbar(list(bit[0][7].keys()),list(bit[0][7].values()),yerr=[list(bit[1][7].values()),list(bit[2][7].values())], label='_nolegend_', marker='o', alpha=1.0, color=color_list[6], linestyle='dashed',markersize=5, capsize=3, ecolor=color_list[6], markerfacecolor=color_list[6], markeredgecolor=color_list[6])
plt.errorbar(list(bit[0][9].keys()),list(bit[0][9].values()),yerr=[list(bit[1][9].values()),list(bit[2][9].values())], label='_nolegend_', marker='o', alpha=1.0, color=color_list[8], linestyle='dashed',markersize=5, capsize=3, ecolor=color_list[8], markerfacecolor=color_list[8], markeredgecolor=color_list[8])

plt.errorbar(list(phase[0][1].keys()),list(phase[0][1].values()),yerr=[list(phase[1][1].values()),list(phase[2][1].values())], label='_nolegend_', marker='s', alpha=1.0, color=color_list[0], linestyle='dotted',markersize=5, capsize=3, ecolor=color_list[0], markerfacecolor=color_list[0], markeredgecolor=color_list[0])
plt.errorbar(list(phase[0][3].keys()),list(phase[0][3].values()),yerr=[list(phase[1][3].values()),list(phase[2][3].values())], label='_nolegend_', marker='s', alpha=1.0, color=color_list[2],linestyle='dotted',markersize=5, capsize=3, ecolor=color_list[2], markerfacecolor=color_list[2], markeredgecolor=color_list[2])
plt.errorbar(list(phase[0][5].keys()),list(phase[0][5].values()),yerr=[list(phase[1][5].values()),list(phase[2][5].values())], label='_nolegend_', marker='s', alpha=1.0, color=color_list[4],linestyle='dotted',markersize=5, capsize=3, ecolor=color_list[4], markerfacecolor=color_list[4], markeredgecolor=color_list[4])
plt.errorbar(list(phase[0][7].keys()),list(phase[0][7].values()),yerr=[list(phase[1][7].values()),list(phase[2][7].values())], label='_nolegend_', marker='s', alpha=1.0, color=color_list[6],linestyle='dotted',markersize=5, capsize=3, ecolor=color_list[6], markerfacecolor=color_list[6], markeredgecolor=color_list[6])
plt.errorbar(list(phase[0][9].keys()),list(phase[0][9].values()),yerr=[list(phase[1][9].values()),list(phase[2][9].values())], label='_nolegend_', marker='s', alpha=1.0, color=color_list[8],linestyle='dotted',markersize=5, capsize=3, ecolor=color_list[8], markerfacecolor=color_list[8], markeredgecolor=color_list[8])
 
ax = plt.gca()
ax.set_xlim([1, 32])
ax.set_ylim([0.975, 1.001])
ax.tick_params(axis='both', which='major', labelsize=12) 
ax.legend(fontsize=11) 
ax.set_xlabel('distance', fontsize=15) 
ax.set_ylabel('Logical fidelity', fontsize=15)
plt.savefig('LogFidvsD.pdf')
plt.show()
```
